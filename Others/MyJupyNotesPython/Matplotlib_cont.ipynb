```python
import matplotlib.pyplot as pyt
import numpy as np
%matplotlib inline
```

```python
numpyArray1 = np.linspace(0,10,15)
```

```python
numpyArray1
```

```python
numpyArray2 = numpyArray1**2
```

```python
numpyArray2
```

```python
(myFig, myAxes) = pyt.subplots()
myAxes.plot(numpyArray1,numpyArray2,"#FF66C0") # color hex code
myAxes.plot(numpyArray2,numpyArray1,"#278BAD",alpha=0.4) # alpha -> transparency option
```

```python
(myFig2, myAxes2) = pyt.subplots()
newNumpyArray=np.linspace(0,40,20)
myAxes2.plot(newNumpyArray,newNumpyArray-5,"#FF66C0",linestyle="--") 
myAxes2.plot(newNumpyArray,newNumpyArray-15,"#FF66C0",marker="o",markersize=8,markerfacecolor="b")
myAxes2.plot(newNumpyArray,newNumpyArray+10,"#FF0060",marker="o",markersize=4,markerfacecolor="b") 
myAxes2.plot(newNumpyArray,newNumpyArray-10,"#366442",linestyle=":",linewidth=4) 
myAxes2.plot(newNumpyArray,newNumpyArray+5,"#B6E1CD",linewidth=5) # linewidth -> kalinlik ayarlama
myAxes2.plot(newNumpyArray,newNumpyArray,"#A46050",linestyle="-.") # linestyle -> cizgi style'i, "--",":" vs.
```
