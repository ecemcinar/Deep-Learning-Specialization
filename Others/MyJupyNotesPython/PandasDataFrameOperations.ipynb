```python
import numpy as np
import pandas as pd
```

```python
 weatherDataFrame=pd.DataFrame({"Istanbul":[30,29,np.nan],"Ankara":[20,np.nan,25],"Izmir":[40,39,38]})
```

```python
weatherDataFrame
```

```python
weatherDataFrame.dropna()
#drop the nan values (from default:"rows")
```

```python
weatherDataFrame
```

```python
weatherDataFrame.dropna(axis=1)
#drop nan values at "columns"
```

```python
 weatherDataFrame2=pd.DataFrame({"Istanbul":[30,29,np.nan],"Ankara":[20,np.nan,25],"Izmir":[40,39,38],"Antalya":[45,np.nan,np.nan]})
```

```python
weatherDataFrame2
```

```python
weatherDataFrame2.dropna(axis=1,thresh=2)
#columns that has more than (or equal) 2 nan val got dropped
```

```python
weatherDataFrame2
```

```python
weatherDataFrame2.fillna(20)
#nan values got filled with 20
```

## groupby method

```python
myDict={"Department":["Software","Sales","Sales","Finance","Finance","Human Res","Law"],
        "Employee":["Micheal","Dwight","Jim","Kevin","Oscar","Toby","Kelly"],
        "Salary":[1000,800,900,600,400,300,500]}
```

```python
salaryDataFrame=pd.DataFrame(myDict)
```

```python
salaryDataFrame
```

```python
depObject=salaryDataFrame.groupby("Department")
```

```python
depObject.count()
```

```python
depObject.mean()
#get mean of salaries for every department
```

```python
depObject.min()
#get min salary of every dep
```

```python
depObject.describe()
```
