```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# jupyter notebook'a ozel, grafiklerin direkt bastirilmasini saglar
# surekli plot.show yazmak zorunda kalmayiz
```

```python
ages = [9,12,18,26,37,43,51,58,79]
weights = [25,32,50,64,70,74,81,54,90]
```

```python
numpyAgeList = np.array(ages)
numpyWeightList = np.array(weights)
```

```python
numpyAgeList
```

```python
plt.plot(numpyAgeList,numpyWeightList,"y")
plt.xlabel("age")
plt.ylabel("weight")
plt.title("Age-Weight Graph")
plt.show()
#(x axis,y axis,color)
```

```python
numpyArray1 = np.linspace(0,10,15)
numpyArray1
```

```python
numpyArray2 =numpyArray1**2
numpyArray2
```

```python
plt.plot(numpyArray1,numpyArray2,"b--")
# "--" koymak kesik cizgili
```

```python
plt.plot(numpyArray1,numpyArray2,"g*-")
# "*-" verilerin oldugu noktalari yildizlayarak gosterir
```

```python
# numb of row, numb of col, which one to print
plt.subplot(1,2,1)
plt.plot(numpyArray1,numpyArray2,"r--")

plt.subplot(1,2,2)
plt.plot(numpyArray2,numpyArray1,"b*-")
```

```python
# creates empty canvas
figure1 = plt.figure()
figureAxes = figure1.add_axes([0.1,0.1,0.6,0.6]) # width-height
figureAxes.plot(numpyArray1,numpyArray2,"g")
figureAxes.set_xlabel("X axis")
figureAxes.set_ylabel("Y axis")
figureAxes.set_title("My Graph")
```

```python
figure2= plt.figure()
axes1 = figure2.add_axes([0.1,0.1,0.7,0.7])
axes2 = figure2.add_axes([0.1,0.3,0.3,0.3])

axes1.plot(numpyArray1,numpyArray2,"r")
axes1.set_xlabel("X axis")
axes1.set_ylabel("Y axis")
axes1.set_title("First Graph")

axes2.plot(numpyArray2,numpyArray1,"g")
axes2.set_xlabel("X axis")
axes2.set_ylabel("Y axis")
axes2.set_title("Second Graph")
```

```python
figure2= plt.figure()
axes1 = figure2.add_axes([0.1,0.1,0.9,0.9])
axes2 = figure2.add_axes([0.2,0.5,0.3,0.3])

axes1.plot(numpyArray1,numpyArray2,"r")
axes1.set_xlabel("X axis for First")
axes1.set_ylabel("Y axis for First")
axes1.set_title("First Graph")

axes2.plot(numpyArray2,numpyArray1,"g")
axes2.set_xlabel("X axis for Second")
axes2.set_ylabel("Y axis for Second")
axes2.set_title("Second Graph")
```

```python
plt.figure()
```

```python
(myfig,myaxes) = plt.subplots(nrows=1,ncols=2)
type(myaxes) # returns (numpy) array

for axes in myaxes:
    axes.plot(numpyArray1,numpyArray2,"y--")
    axes.set_xlabel("x axis")
    axes.set_ylabel("y axis")
    axes.set_title("Title")
    
plt.tight_layout()
```

```python
newFig = plt.figure()

newAxes = newFig.add_axes([0.1,0.1,0.9,0.9])

newAxes.plot(numpyArray1,numpyArray2,label="numpy**2")
newAxes.plot(numpyArray1,numpyArray1**3,label="numpy**3")
newAxes.legend()
```

```python
newFig2 = plt.figure(figsize=(4,4))

newAxes2 = newFig2.add_axes([0.1,0.1,0.9,0.9])

newAxes2.plot(numpyArray1,numpyArray2,label="numpy**2")
newAxes2.plot(numpyArray1,numpyArray1**3,label="numpy**3")
newAxes2.legend(loc=1)
```

```python
newFig2 = plt.figure(dpi=100)
#dpi -> cozunurluk icin
newAxes2 = newFig2.add_axes([0.1,0.1,0.9,0.9])

newAxes2.plot(numpyArray1,numpyArray2,label="numpy**2")
newAxes2.plot(numpyArray1,numpyArray1**3,label="numpy**3")
newAxes2.legend(loc=1)
```
